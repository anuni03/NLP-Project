```python
import torch
from transformers import AutoTokenizer,AutoModelForCausalLM
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model_name="gpt2-large"

tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
input_text="i am software engineer"
max_length=128

input_ids= tokenizer(input_text,return_tensors="pt")
```

```python
input_ids
```

```python
input_ids=input_ids['input_ids'].to(device)
```

```python
output=model.generate(input_ids,max_length=max_length,num_beams=5,do_sample=False)
```

```python
output
```

```python
print(tokenizer.decode(output[0]))
```

```python
output=model.generate(input_ids,max_length=max_length,num_beams=5,do_sample=False,no_repeat_ngram_size=2)
tokenizer.decode(output[0])
```

## NUCLEAR SEARCH


```python
input_text="Hello, My name is Anushka, I like to Code"
max_length=128

input_ids= tokenizer(input_text,return_tensors="pt")
input_ids=input_ids['input_ids'].to(device)


output=model.generate(input_ids,max_length=max_length,do_sample=False,top_p=0.9)
tokenizer.decode(output[0])
```
